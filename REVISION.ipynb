```python
months = ['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec']
customer1  = np.arange(2,26,2)
customer2 = [1,3,3,5,5,7,23,12,8,9,11,13]
```

```python
customer1
```

```python
customer2 
```

```python
plt.bar(months,customer2, color = 'red')
plt.xlabel('Month')

plt.ylabel('Customer2')
plt.show()
```

```python
bar_size =0.4
month = np.arange(12)
plt.bar(month,customer1,bar_size, color = 'blue', label ='customer1')
plt.bar(month+bar_size,customer2,bar_size, color = 'green',label = 'customer2')
plt.xlabel('Months')
plt.xticks(month + (bar_size)/12,months )
plt.ylabel('consumption')
plt.title('Electricity Usage')
plt.show()
```
